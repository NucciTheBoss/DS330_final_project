```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
df = pd.read_csv('/content/drive/My Drive/spotify_data/sampled_spotify.csv')
```

Take three days before each event and three days after. Look at quantatative change in the beat, danceability, and energy. Maybe add some cool colors.

```python
df_firstevent = df[(df['date'] >= '2018-07-15') & (df['date'] <= '2018-07-17')]
df_secondevent = df[(df['date'] >= '2018-07-25') & (df['date'] <= '2018-07-27')]
df_thirdevent = df[(df['date'] >= '2018-08-05') & (df['date'] <= '2018-08-07')]
df_fourthevent = df[(df['date'] >= '2018-09-10') & (df['date'] <= '2018-09-12')]
```

```python
grouped_df = df_firstevent.groupby("date")
mean_df1 = grouped_df["beat_strength","danceability","energy"].mean()
mean_df1 = mean_df1.reset_index()
mean_df1['Event'] = "Jeff Bezos Wealth"
mean_df1['Date_Order'] = "0"
for i in range(len(mean_df1)):
  if mean_df1['date'][i] == "2018-07-15":
    mean_df1["Date_Order"][i] =1
  elif mean_df1['date'][i] == "2018-07-16":
    mean_df1["Date_Order"][i] =2
  elif mean_df1['date'][i] == "2018-07-17":
    mean_df1["Date_Order"][i] =3


print(mean_df1)

grouped_df = df_secondevent.groupby("date")
mean_df2 = grouped_df["beat_strength","danceability","energy"].mean()
mean_df2 = mean_df2.reset_index()
mean_df2['Event'] = "Facebook Sell Off"
mean_df2['Date_Order'] = "0"
for i in range(len(mean_df2)):
  if mean_df2['date'][i] == "2018-07-25":
    mean_df2["Date_Order"][i] =1
  elif mean_df2['date'][i] == "2018-07-26":
    mean_df2["Date_Order"][i] =2
  elif mean_df2['date'][i] == "2018-07-27":
    mean_df2["Date_Order"][i] =3

print(mean_df2)

grouped_df = df_thirdevent.groupby("date")
mean_df3 = grouped_df["beat_strength","danceability","energy"].mean()
mean_df3 = mean_df3.reset_index()
mean_df3['Event'] = "Saudi Arabia and Canadian Diplomats"
mean_df3['Date_Order'] = "0"

for i in range(len(mean_df3)):
  if mean_df3['date'][i] == "2018-08-05":
    mean_df3["Date_Order"][i] =1
  elif mean_df3['date'][i] == "2018-08-06":
    mean_df3["Date_Order"][i] =2
  elif mean_df3['date'][i] == "2018-08-07":
    mean_df3["Date_Order"][i] =3

print(mean_df3)

grouped_df = df_fourthevent.groupby("date")
mean_df4 = grouped_df["beat_strength","danceability","energy"].mean()
mean_df4 = mean_df4.reset_index()
mean_df4['Event'] = "Hurricane Florence"
mean_df4['Date_Order'] = "0"

for i in range(len(mean_df4)):
  if mean_df4['date'][i] == "2018-09-10":
    mean_df4["Date_Order"][i] =1
  elif mean_df4['date'][i] == "2018-09-11":
    mean_df4["Date_Order"][i] =2
  elif mean_df4['date'][i] == "2018-09-12":
    mean_df4["Date_Order"][i] =3

print(mean_df4)


result = pd.concat([mean_df1, mean_df2,mean_df3,mean_df4], axis=0)
result
```

```python
fig1 = px.scatter(result, x=result["beat_strength"], y=result["date"], color="Event", symbol="Event")
fig1.update_layout( title={
        'text': "Beat Strength",
        'y':.95,
        'x':0.45,
        'xanchor': 'center',
        'yanchor': 'top'},
        font=dict(
            size=12,
        ))
fig1.update_traces(marker=dict(size=10,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig1.show()
```

```python
fig2 = px.scatter(result, x=result["danceability"], y=result["date"], color="Event", symbol="Event")
fig2.update_layout( title={
        'text': "Danceability",
        'y':.95,
        'x':0.45,
        'xanchor': 'center',
        'yanchor': 'top'},
        font=dict(
            size=12,
        ))
fig2.update_traces(marker=dict(size=10,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig2.show()
```

```python
fig3 = px.scatter(result, x=result["energy"], y=result["date"], color="Event", symbol="Event")
fig3.update_layout( title={
        'text': "Energy",
        'y':.95,
        'x':0.45,
        'xanchor': 'center',
        'yanchor': 'top'},
        font=dict(
            size=12,
        ))
fig3.update_traces(marker=dict(size=10,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig3.show()
```

```python


fig4 = px.scatter_3d(result, x=result["beat_strength"], y=result["danceability"], z=result["energy"],
                    color='Event', symbol='date')

fig4.update_layout( title={
        'text': "3-D Interactions of Beat, Danceability, and Energy",
        'y':.95,
        'x':0.45,
        'xanchor': 'center',
        'yanchor': 'top'},
        font=dict(
            size=12,
        ))


fig4.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig4.update_layout(scene = dict(xaxis = dict(nticks=4, range=[.51,.55],),yaxis = dict(nticks=4, range=[.65,.67],),zaxis = dict(nticks=4, range=[.62,.64],)))





fig4.show()
```
